# The Office
I like watching light-hearted TV shows while having lunch and dinner. This led me to come up with a mini project to generate two random numbers which will give me valid values for season number and episode number for the famous TV series - The Office.

So, now we know that different seasons have different number of episodes. Let's generate the first random number which will give us the season number.

```python
import random
```

```python
season_selector = random.randint(1,9)
```

Now, depending on which season is selected, we will need to assign a random number to get a valid episode number for the given season.

```python
no_of_episodes = {
    1: random.randint(1,6),
    2: random.randint(1,22),
    3: random.randint(1,25),
    4: random.randint(1,19),
    5: random.randint(1,28),
    6: random.randint(1,26),
    7: random.randint(1,26),
    8: random.randint(1,24),
    9: random.randint(1,25),
}
```

```python
print('Season:', season_selector)
print('Episode:', no_of_episodes[season_selector])
```
